```python
import numpy as np
from scipy.stats import geom
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(17)
```

```python
# Example 1 1000 random samples with numpy
```

```python
p = 0.125
```

```python
samples = np.random.geometric(p, 1000)
```

```python
print(samples)
```

```python
# Example 2 Calc Mean, Var, STD
```

```python
np.mean(samples)
```

```python
np.var(samples)
```

```python
np.std(samples)
```

```python
# Example 3 Histograms
```

```python
plt.figure(figsize=(15,6))
plt.hist(samples, bins=np.arange(1, 53) -0.5, edgecolor='black', rwidth=0.8)
plt.xlabel('Packes Opened')
plt.ylabel('Frequency')
plt.title('Histogram of Geo Data')
plt.xticks(np.arange(1,53))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Example 4 PMF
```

```python
x = np.arange(1, max(samples) + 1)
```

```python
pmf = geom.pmf(x, p)
```

```python
plt.hist(samples, bins=np.arange(0.5, max(samples) + 1.5), density=True, alpha=0.6, color='green', edgecolor='black', rwidth=0.8)
plt.xlabel('Number of Trials')
plt.ylabel('Probability')
plt.title('PMF of Geometric Distribution')
plt.grid(True)
plt.show()
```

```python
# Example 5 CDF
```

```python
cdf = geom.cdf(x, p)
```

```python
plt.step(x, cdf, where='post', color='blue')
plt.xlabel('Number of Trials')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Geometric Distribution')
plt.grid(True)
plt.show()
```
